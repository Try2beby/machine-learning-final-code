```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

import os

dataDir = "../../data/ml-1m/"
```

```python
# Load data
ratings = pd.read_csv(
    dataDir + "ratings.dat",
    sep="::",
    names=["UserID", "MovieID", "Rating", "Timestamp"],
    engine="python",
)
users = pd.read_csv(
    dataDir + "users.dat",
    sep="::",
    names=["UserID", "Gender", "Age", "Occupation", "Zip-code"],
    engine="python",
)
movies = pd.read_csv(
    dataDir + "movies.dat",
    sep="::",
    names=["MovieID", "Title", "Genres"],
    engine="python",
    encoding="ISO-8859-1",
)

text = """
    * Action
    * Adventure
    * Animation
    * Children's
    * Comedy
    * Crime
    * Documentary
    * Drama
    * Fantasy
    * Film-Noir
    * Horror
    * Musical
    * Mystery
    * Romance
    * Sci-Fi
    * Thriller
    * War
    * Western
"""
lines = text.splitlines()
genres = [line.strip(" *") for line in lines if line.strip()]

# Merge dataframes
data = pd.merge(pd.merge(ratings, users), movies)


def encode_genres(df):
    """Encode genres into a binary vector."""
    genres = df["Genres"].str.get_dummies(sep="|")
    return pd.concat([df, genres], axis=1).drop("Genres", axis=1)


# encode genres
data = encode_genres(data)
# drop genres
# data = data.drop("Genres", axis=1)

# merge columns in genres to one column, new column is a list of all values
# data["Genres"] = data[genres].values.tolist()

# drop columns in genres
# data = data.drop(genres, axis=1)

# Encode categorical variables
le = LabelEncoder()
data["Gender"] = le.fit_transform(data["Gender"])
# data["Title"] = le.fit_transform(data["Title"])

# extract year, month, day from timestamp
data["Year"] = pd.to_datetime(data["Timestamp"], unit="s").dt.year
data["Month"] = pd.to_datetime(data["Timestamp"], unit="s").dt.month
data["Day"] = pd.to_datetime(data["Timestamp"], unit="s").dt.day
# drop timestamp
data = data.drop("Timestamp", axis=1)

# save data
data.to_csv(os.path.join(dataDir, "data.csv"), index=False)
```
